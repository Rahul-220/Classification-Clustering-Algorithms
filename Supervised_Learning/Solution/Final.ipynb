```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import graphviz
from IPython.display import display
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.tree import export_graphviz
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score

nba= pd.read_csv('nba_stats.csv')
data = nba[['G', 'FG%', '3P%', '2P%', 'eFG%', 'FT%', 'ORB', 'DRB', 'AST', 'STL','BLK', 'PTS']]
target = nba['Pos']
train_feature, test_feature, train_class, test_class = train_test_split(
    data, target, test_size=0.2, stratify=target, random_state=0)


# Define and train SVM model
svm = SVC(random_state=0,kernel='linear')
svm.fit(train_feature, train_class)
accuracy = svm.score(test_feature, test_class)
print("Training set score: {:.2f}".format(svm.score(train_feature, train_class)))
print("Test set accuracy: {:.2f}".format(accuracy))

# Confusion Matrix:
prediction = svm.predict(train_feature)
print("Confusion matrix for training:")
print(pd.crosstab(train_class, prediction, rownames=['True'], colnames=['Predicted'], margins=True))

prediction = svm.predict(test_feature)
print("Confusion matrix for validation:")
print(pd.crosstab(test_class, prediction, rownames=['True'], colnames=['Predicted'], margins=True))


# Dummy set Accuracy
dummy = pd.read_csv('dummy_test.csv')
data_1 = dummy[['G', 'FG%', '3P%', '2P%', 'eFG%', 'FT%', 'ORB', 'DRB', 'AST', 'STL','BLK', 'PTS']]
target_1 = dummy['Pos']

accuracy = svm.score(data_1, target_1)
print("Test set accuracy: {:.2f}".format(accuracy))


# Confusion Matrix:
prediction = svm.predict(data_1)
print("Confusion matrix for training:")
print(pd.crosstab(target_1, prediction, rownames=['True'], colnames=['Predicted'], margins=True))


# Cross-validation
nba= pd.read_csv('nba_stats.csv')
data = nba.drop(['Pos','Age','GS','MP','FG','TOV','FTA','FT','PF','PTS','2PA','ORB'], axis=1)
target = nba['Pos']
scores = cross_val_score(svm, data, target, cv=10)
print("Cross-validation scores: {}".format(scores))
print("Average cross-validation score: {:.2f}".format(scores.mean()))
  
```
